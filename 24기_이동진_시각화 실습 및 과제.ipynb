## visualization with Matplotlib

```python
import numpy as np

from matplotlib.colors import Normalize
from matplotlib import colormaps
from matplotlib.markers import MarkerStyle
from matplotlib.text import TextPath
```

#### 1번 시각화

```python
zipped_data = zip(df['carat'], df['cut'], df['color'], df['clarity'], df['price'])

cmap = plt.colormaps["Purples"]
fig, ax = plt.subplots()
fig.suptitle("Throwing success", size=14)
for carat, cut, color, clarity, price in zipped_data:
    ax.plot(color,
            price,
            marker='*',
            color=cmap(carat))
fig.colorbar(plt.cm.ScalarMappable(norm=Normalize(0, 5),
                                   cmap=cmap),
             ax=ax, label="carat")
ax.set_xlabel("color")
ax.set_ylabel("price")
ax.set_ylim(0, 5000)

plt.show()
```

2. 1, 2, 3번 시각화 중 하나를 골라 1번 문항에서 제안한 대로 개선하세요. (300자 이내) 어떤 파이썬 라이브러리나 데이터 시각화 툴을 이용할지는 자유입니다. matplotlip, seaborn, plotly 공식문서 혹은 (태블로를 사용하고 싶다면) 태블로 시 블로그를 참고하세요.

1번 시각화를 개선하기 위해 앞선 문항에서 제시한 바와 같이 점의 색상의 변화가 뚜렷하지 않아 캐럿에 따른 다이아몬드의 색상의 분포를 알기 어렵기에 다른 색상 팔레트를 사용하여 뚜렷한 색상대비를 갖도록 하여 이를 개선하였다. 원래는 점의 색상과 함께 점 크기 또한 사용하여 나타내려고 하였으나 지나치게 실행하는데에 오래 걸려 점 크기는 사용하지 않았다. 또한 제목이 시각화 내용을 잘 설명하지 못한다고 판단하여서 Diamond Price by Color and Carat이라는 제목으로 수정하였다.

```python
zipped_data = zip(df['carat'], df['cut'], df['color'], df['clarity'], df['price'])

# 색상 팔레트를 'viridis'로 변경
cmap = plt.get_cmap('viridis')
fig, ax = plt.subplots()

# 제목을 "Diamond Price by Color and Carat"으로 변경
fig.suptitle("Diamond Price by Color and Carat", size=14)

for carat, cut, color, clarity, price in zipped_data:
    ax.plot(color,
            price,
            marker='*',
            color=cmap(carat))
fig.colorbar(plt.cm.ScalarMappable(norm=Normalize(0, 5),
                                   cmap=cmap),
             ax=ax, label="carat")
ax.set_xlabel("color")
ax.set_ylabel("price")
ax.set_ylim(0, 5000)

plt.show()
```
